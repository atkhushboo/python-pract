```python
#Access modifiers == Encapsulation
#private acess modifire [--]

class Person:

    def __init__(self,name,age):
        self.__name= name
        self.__age = age

    def display_info(self):
        return f"Name is {self.__name} and age is {self.__age}"

p1 = Person("khush",12)
# print(p1.__age) .....not Acceptanle
# dir(p1)....directry of p1 object to check
# print(p1._Person__name) .....Acceptable but not a good way to acces

print(p1.display_info()) #..........this is a good way to access private varibles
```

```python
#protected [-].....internal uses .....use with drived class 
# it can only use in drived class

class Person:

    def __init__(self,name,age):
        self._name= name
        self._age = age

class Student(Person):              #inherit Person class ....  #..This is a drive class of {Person}
    def __init__(self,name,age):
        super().__init__(name,age)
    
    def display_detail(self):
        return f"Name is:- {self._name} and the age is:- {self._age}"
    
s1 = Student("Protected",20)
print(s1.display_detail())
```
