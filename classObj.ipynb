```python
class Form:

    total_form=0
    def __init__(self,name,last_name,age):
        self.name= name
        self.last_name = last_name
        self.__age = age
        Form.total_form +=1
    


    # method that return full-name
    def full_name(self):
        return f"{self.name} {self.last_name}"
    
    # **************end*******
    @staticmethod
    def static_learn():
        return "Static method acces class & object without (self(object data))"
    
    @property
    def age(self):
        return self.__age
    
    # ******encapsulation*** to private attribute******** 
    def get__age(self):
        return self.__age,"!"
    
    # **********end******
    # ********polymorphism*********
    
    def topic(self):
        return "encapsulation-topic"
    
    # **************end**************

    # **************inheritance**********
class gender_sex(Form):
    def __init__(self,name,last_name,__age,gender):

        super().__init__(name,last_name,__age)
        self.gender=gender

    # ********polymorphism*********
    
    def topic(self):
        return "inheritance-topic"
    
    # *********end************
    
sex = gender_sex("khushi","kumari",19,"female")
print(sex.full_name(),sex.get__age(),sex.gender)
print(sex.topic())

# age is private attribute that's why we use get_age

# ****************end***********
    
    
first_form=Form("khushboo","kumari",20)
# first_form.age=19
print(first_form.topic())
print(Form.total_form)
print(first_form.static_learn())
print(Form.static_learn())
print(first_form.age)

# print(first_form.name)
# print(first_form.last_name)
# print(first_form.age)
# print(first_form.age())


# *************end*******
# method and self
# print(first_form.full_name())
```

```python
# bike class
class Bike:
    
    def __init__(self,brand,model):
        self.brand = brand
        self.__model = model

    # **********encapsulate**********hide data*****

    # def get_model(self):
    #     return self.__model
     
    # ************never change only for read ************    
    @property
    def modal(self):
        return self.__model

bike_one = Bike("hero","splender")
bike_one.model="activa"
print(bike_one.brand)
print(bike_one.modal())
```

```python
import math

class   Circle:

    def __init__(self,radius):
        self.radius = radius

    def get_area(self):
        return "Area of circle:-",math.pi*self.radius*self.radius
    
    def get_peramiter(self):
        return "Perameter of Circle:-",2*math.pi*self.radius
    
c_one = Circle(5)
print(c_one.get_area())
print(c_one.get_peramiter())
```

```python
# create student class that take name and marks of 3 subject as a argument constructor. 
# then create a method to print average

class Student:
    def __init__(self,name,marks):
        self.name=name
        self.marks = marks

    @staticmethod
    def collage_name():
        return "Panjab University"
    
    def average(self):
        count =0
        arr =[]
        for i in range(len(self.marks)):
            arr.append(self.marks[i])
            count=count+1
            avg= sum(arr)/count
            
        return "the average number of",self.name,"is :-",avg
        
s1= Student("khushi",[90,45,85])
print(s1.collage_name())
# print(s1.name,s1.marks)
print(s1.average())

s2 = Student("khushbu",[98,99,97])
# print(s2.name,s2.marks)
print(s2.average())
```
