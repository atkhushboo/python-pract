```python
class student:

    def __init__(self,name,age,course):
        self.name=name
        self.age=age
        self.course=course

class_one = student("Khushboo",20,"MCA")
print(class_one.name,class_one.age,class_one.course)

class_two =student("Atul",21,"MCA")
print(class_two.name,class_two.age,class_two.course)

class_three =student("Khushi",19,"BCA")
print(class_three.name,class_three.age,class_three.course)
```
