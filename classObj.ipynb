```python
# create student class that take name and marks of 3 subject as a argument constructor. 
# then create a method to print average

class Student:
    def __init__(self,name,marks):
        self.name=name
        self.marks = marks

    @staticmethod
    def collage_name():
        return "Panjab University"
    
    def average(self):
        count =0
        arr =[]
        for i in range(len(self.marks)):
            arr.append(self.marks[i])
            count=count+1
            avg= sum(arr)/count
            
        return "the average number of",self.name,"is :-",avg
        
s1= Student("khushi",[90,45,85])
print(s1.collage_name())
# print(s1.name,s1.marks)
print(s1.average())

s2 = Student("khushbu",[98,99,97])
# print(s2.name,s2.marks)
print(s2.average())
```

```python
# create account class with 2 attribute balance and account number:
# create method for debit and credit and print balance:
    
class Account:
    def __init__(self,balance,account_no):
        self.balance=balance
        self.account_no=account_no
    
    def credit(self,amount):
        self.balance = self.balance + amount
        print(f"Credit in acount: {amount}")
        print(f"Updated balance: {self.balance}")
        
    def debit(self,amount):
        self.balance = self.balance - amount
        print("Debit from account:" ,amount)
        print("Current balance is: ",self.balance)
        
acc =Account(100,56783452)
print("Account Number is:" ,acc.account_no)
print("balance is: ",acc.balance)
acc.credit(5000)
acc.debit(2000)
```

```python
# private attribute and methods

class Login:

    def __init__(self,userName,password):
        self.userName= userName
        self.__password = password    #_ _ is private attribte
    
    def __secure_data(self):          #__ is private method 
        return "password is:-",self.__password
    
    def show_secure_data(self):        
        return self.__secure_data()       # call private method to call __password and print

userOne = Login("khush@123",70075)
print(userOne.userName)
print(userOne.show_secure_data())
print(userOne._Login__password) # hack type  but not recomended...........
```

```python
# inheritance : - inherit / drive the properties and method to the one class to another classs/

# single inheritance:-

class Mobile:
    
    @staticmethod
    def mobile_price():
        return "65000"
    
class Color(Mobile):     #single inheritance

    def __init__(self,colors):
        self.colors = colors

class Brand(Color): #multi-level inheritance.......
    
    def __init__(self,name,colors):
        super().__init__(colors)   #use to inherite /acces the method of parent class.......
        self.name=name

class Country(Brand):

    @staticmethod
    def country():
        return "USA"
    

# mobile_one = Color ("White")
# print(mobile_one.mobile_price())
# print(mobile_one.colors)  
# b1 = Brand("iphone","White")
# print(b1.mobile_price())
# print(b1.colors)
cn = Country("iphone","white")
print(cn.name,cn.colors,cn.mobile_price(),cn.country())
```

```python
# multiple inheritance

class First:
    
    @staticmethod
    def section_A():
        return "Section -A"

class Second:
    
    @staticmethod
    def section_B():
        return "Section -B"
    
class Third:

    @staticmethod
    def student():
        return "Khushboo and Khushi"

class College(First,Second,Third):
    
    @staticmethod
    def college():
        return "Panjab university"

c1 = College()
print(c1.section_A(),c1.section_B(),c1.student(),c1.college())

```
