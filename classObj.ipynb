```python
class student:

    def __init__(self,name,age,course="null"):
        self.name=name
        self.age=age
        self.course=course

# instance or object
class_one = student("Khushboo",20,"MCA")
print(class_one.name,class_one.age,class_one.course)

class_two =student("Atul",21,"MCA")
print(class_two.name,class_two.age,class_two.course)

class_three =student("Khushi",19)
print(class_three.name,class_three.age)
```

```python
class Form:

    total_form=0
    def __init__(self,name,last_name,age):
        self.name= name
        self.last_name = last_name
        self.__age = age
        Form.total_form +=1

    # method that return full-name
    def full_name(self):
        return f"{self.name} {self.last_name}"
    
    # **************end*******
    @staticmethod
    def static_learn():
        return "Static method acces class & object without (self(object data))"
    
    # ******encapsulation*** to private attribute******** 
    def get__age(self):
        return self.__age,"!"
    
    # **********end******
    # ********polymorphism*********
    
    def topic(self):
        return "encapsulation-topic"
    
    # **************end**************

    # **************inheritance**********
class gender_sex(Form):
    def __init__(self,name,last_name,__age,gender):

        super().__init__(name,last_name,__age)
        self.gender=gender

    # ********polymorphism*********
    
    def topic(self):
        return "inheritance-topic"
    
    # *********end************
    
sex = gender_sex("khushi","kumari",19,"female")
print(sex.full_name(),sex.get__age(),sex.gender)
print(sex.topic())

# age is private attribute that's why we use get_age

# ****************end***********
    
    
first_form=Form("khushboo","kumari",20)
print(first_form.topic())
print(Form.total_form)
print(first_form.static_learn())
print(Form.static_learn())

# print(first_form.name)
# print(first_form.last_name)
# print(first_form.age)
# *************end*******
# method and self
# print(first_form.full_name())
```

```python
# bike class
class Bike:
    
    def __init__(self,brand,model):
        self.brand = brand
        self.__model = model

    # **********encapsulate**********hide data*****

    def get_model(self):
        return self.__model    

bike_one = Bike("hero","splander")
print(bike_one.brand)
print(bike_one.get_model())
```
