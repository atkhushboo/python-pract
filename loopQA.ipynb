```python
# Print list in reverse order using a loop

org_arr=["k","h","u","s","h"]
rev_arr=[]

for i in range(len(org_arr)-1,-1,-1):
    rev_arr.append(org_arr[i])
    
print(rev_arr)
```

```python
# Find the factorial of a given number

fact=1
for i in range(1,5+1):
    fact=fact*i
print(fact)

# ***************using function***********

def fact_num(n):    
    result =1

    for i in range(1,n+1):
        result= result*i
    
    return result

output=fact_num(7)
print("The factorial number is :-",output)
```

```python
# Find largest and smallest digit in a number

def min_max(n):
    
    num=str(n)
    arr=[]
    print("Check the number",type(num))
    
    for i in num:
        arr.append(i)
        print(arr)

    return " Largest and Smalles" ,max(arr),min(arr)

output = min_max(2317693)
print(output)
```

```python
# Exercise 12: Display Fibonacci series up to 10 terms....

n_num = 10 
first ,second =0,1

count = 0

while count<n_num:

    print(first)

    n_term = first+second
    first = second
    second = n_term 
    count= count+1


```
