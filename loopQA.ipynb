```python
#  Calculate sum of all numbers from 1 to a given number

i=0
sum=0

while i<10:
    sum=sum+i
    i=i+1
    print(sum)
```

```python
# Print list in reverse order using a loop

org_arr=["k","h","u","s","h"]
rev_arr=[]

for i in range(len(org_arr)-1,-1,-1):
    rev_arr.append(org_arr[i])
    
print(rev_arr)
```
