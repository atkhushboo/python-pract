```python
# Multiply Items in List
# Write a Python program to multiply all the items in a list

def fun(n):
   mul =1

   for i in n:
      # print(i)
      mul= mul*i
      # print(mul)
    
   return "Multiplication of all items",mul

print(fun([1,2,3,4,5,6]))
```

```python
# Get Largest and smallest Number in List
# Write a Python program to get the largest and smallest number from a list.

def larger(num):
     
    ma=max(num)
    mi=min(num)
       
    return "Largest Num is :-",ma,"Smallest num is:-",mi
print(larger([45,68,77,12,34]))
```

```python
# Count Strings with Same Start and End
# Write a Python program to count the number of strings from a given list of strings,
# The string length is 2 or more and the first and last characters are the same.

def check(item):

    return item[0]==item[-1]
# print(check("abca"))

def count_item(arr):
     
    print(arr)
    count=0
    for i in arr:
        # print(i)
        print(check(i))

        if check(i):
            count=count+1
    
    return "Total Characters are the same from First and Last:- ",count

print(count_item(["abc","aabb","abca","khk","12121","23232"]))
```
