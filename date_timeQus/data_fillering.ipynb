```python
# Qustion 01 :-  Data Filtering + Conditions......
#             - Ek list hai users ki (dict format me), sirf un users ko return karo:
#             - jinka is_active = True
#             - aur last_login 30 days ke andar hai
#             - null values skip karo
```

```python
from datetime import datetime
def data_filter():
    user_list=[
         {
        "id": 1,
        "name": "Amit",
        "email": "amit@gmail.com",
        "is_active": True,
        "last_login": "2026-03-20"
      },
      {
        "id": 2,
        "name": "Neha",
        "email": "neha@gmail.com",
        "is_active": False,
        "last_login": "2026-03-25"
      },
      {
        "id": 3,
        "name": "Rahul",
        "email": "rahul@gmail.com",
        "is_active": True,
        "last_login": "2026-02-10"
      },
      {
        "id": 4,
        "name": "Priya",
        "email": "priya@gmail.com",
        "is_active": True,
        "last_login": "2026-03-01"
      },
      {
        "id": 5,
        "name": "Karan",
        "email": "karan@gmail.com",
        "is_active": True,
        "last_login": None
      },
      {
        "id": 6,
        "name": "Sneha",
        "email": "sneha@gmail.com",
        "is_active": True,
        "last_login": "2026-03-26"
      }
    ]

    result=[]
    today=datetime.today()

    for users in user_list:
        if users["is_active"] and users["last_login"] is not None:
            
            date=datetime.strptime(users["last_login"],"%Y-%m-%d")
            diff = today - date

            if diff.days<=30:
                result.append(users)
    return result

print(data_filter())
```

```python
# 🧠 Problem Simple Language me
# 👉 Data aisa aa sakta hai:
# key missing ❌
# value None ❌
# 👉 hume:
# error nahi aana chahiye
# invalid data skip karna hai
# sirf valid data process karna hai
```

```python
data= [{"id": 1, "value": 10},
        {"id": 2},
        {"id": 3, "value": None},
        {"id": 4, "value": 25},
        {"value": 30},
        {"id": 6, "value": 0}

    ]
def invalid_data_check(data):
    result=[]
    for item in data:
        if "id" not in item or "value" not in item:
            continue
        if item["value"] is None:
            continue
        result.append(item)
    return result
print(invalid_data_check(data))
```
